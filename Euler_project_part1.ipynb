```julia
#problem 21
function divisor(n::Int)
    z = Int[]
    for i in 1:n-1
        if n%i == 0 
            push!(z,i)
        end
    end
        if sum(z) == n
            return nothing
    else
        return sum(z)
    end
end
divisor(496)
```

```julia
function amicable(n,m)
    if divisor(n) == m && divisor(m)==n
        return true
    else 
        return false
    end
end
amicable(5020,5564)
```

```julia
m33 = []
for i in 1:10000
    for j in 1:10000
        if amicable(i,j) && !(i in m33)
            push!(m33,i)
        end
    end
end
m33
sum(m33)
```
